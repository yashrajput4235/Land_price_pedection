```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model
```

Now read the csv file

```python
df = pd.read_excel('/content/price_zameen.xlsx')
df
```

Now I have the data frames now i just scatter plot to know the distribution of the data frame


```python
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US $)')
plt.scatter(df.area,df.price,color='red',marker='+')

```

Now applying linear regression model

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

Now taking an example predict the price of home whose area is 3300 sqauare feet


```python
reg.predict([[3300]])
```

Now finding the value of cofficient that is slope
  

```python
reg.coef_
```

Now seeing the intercept


```python
reg.intercept_
```

The area are given now we have to predict prices


```python
d = pd.read_excel('/content/zameen_final.xlsx')
d
```

```python
p=reg.predict(d)
p
```

```python
d['price']=p
d
```

Now exporting the prices to zameen_final because in that only area is given

```python
d.to_excel('zameen_final.xlsx',index=False)
```

Now isual representaion of linear regression line
  

```python
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US $)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
